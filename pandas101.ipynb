##### loading data

```python
import pandas as pd
```

```python
df = pd.read_csv("pokemondata.csv")
df
```

```python
# headers
print(df.columns)
```

```python
# read each column
print("name only\n", df["Name"])
print("\n\nname until 5th row\n", df["Name"][0:5])
print("\n\nname + type + hp\n", df[["Name","Type 1","HP"]])
```

```python
# read each row
print(df.head(4)) # print first four rows
print(df.iloc[1]) # integer location - print specific row
```

```python
# read specific location (col + row)
print(df.iloc[3][6])
```

```python
# read each row (alternative)
for index, row in df.iterrows():
    print(index, row["Name"])

for index, row in df.iterrows(): # does the same thing as above code
    print(index, row[1])
```

```python
# specific conditions met (all pokemon with attack > 50)
df.loc[df["Attack"] > 50]
```

#### High level data -> describe AND sorting

```python
df.describe()
```

```python
df.sort_values(["HP"], ascending = True)
```

```python
# sorting multiple columns
df.sort_values(['Speed', 'Name'], ascending = [False, True])
```

#### Making changes to data


```python
df["Total"] = df["HP"] + df["Attack"] + df["Defense"] + df["Speed"]
```

```python
df.sort_values(["Total"], ascending = True)
```

```python
## Dropping column
df.drop(columns=["Total"])
```

```python
# adding Total but with integer location - iloc
df["Total"] = df.iloc[:, 4:10].sum(axis=1)
df
```

```python
# moving the Total column to the middle before HP
cols = list(df.columns)
df = df[cols[0:4] + [cols[-1]] + cols[4:12]]
df
```

#### Saving as CSV

```python
df.to_csv("modified.csv")
```

```python
# without indexes
df.to_csv("modified_no_index.csv", index = False)
```

#### Filtering Data

```python
df
```

```python
# AND = &, OR = |
df.loc[ (df["Type 1"] == "Fire") & (df["Attack"] > 50)]
```

```python
df_filtered_grass_poison = df.loc[ (df["Type 1"] == "Grass") | (df["Type 2"] == "Poison") ]
df_filtered_grass_poison
```

```python
df_filtered_grass_poison.to_csv("grass_poison.csv")
```

#### filter out names that include "Mega"

```python
df.loc[df["Name"].str.contains("Mega")]
```

```python
# droppping all MEGAS out --> ~
df_no_mega = df.loc[~df["Name"].str.contains("Mega")]
df_no_mega
```

```python
df_no_mega.to_csv("no_megas.csv")
```

#### Regular Expressions

```python
import re
```

```python
df.loc[df["Type 1"].str.contains("fire|grass", flags=re.I, regex = True)] # flags = ignore case 
```

```python
# name starts with PI ---- 
## pi anywhere in the name
df.loc[df["Name"].str.contains("pi[a-z]*", flags=re.I, regex = True)]
```

```python
## pi in the beggining of the sentence
df.loc[df["Name"].str.contains("^pi[a-z]*", flags=re.I, regex = True)]
```

### Conditional Changes 

```python
# change type into another word

new_df = df

new_df.loc[new_df["Type 1"] == "Fire", "Type 1"] = "Heat"
new_df
```

```python
# change legendary status

new_df.loc[new_df["Type 1"] == "Heat", "Legendary"] = True
new_df
```

```python
# change multiple columns at once

new_df.loc[new_df["HP"] > 90, ["Type 1", "Legendary"]] = ["Very COOL", True]
new_df.loc[new_df["HP"] > 90]
```

### GROUP BY

```python
new_df = pd.read_csv("pokemondata.csv")
new_df
```

```python
new_df = new_df.drop(columns="Name")
new_df = new_df.drop(columns="Type 2")
new_df = new_df.drop(columns="Legendary")
```

```python
new_df
```

```python
new_df.groupby(["Type 1"]).mean().sort_values("HP", ascending = False)
```

```python
new_df["count"] = 1
new_df.groupby("Type 1").count()["count"].sort_values(ascending=False)
```
